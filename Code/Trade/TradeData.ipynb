```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trade-data.csv',encoding='latin1')
```

```python
df.head()
print(df.shape)
df.head(7072)
```

```python
df['Value']
```

```python
indicators = ['Total merchandise imports - monthly','Total merchandise exports - monthly','Commercial services exports of selected economies monthly','Commercial services imports of selected economies monthly']
```

```python
for i in indicators:
    print(i)
```

```python
t = df['Reporting Economy'].unique().tolist()
print(t)
```

```python
list1 = ['Total merchandise imports - monthly']
listofcountry = ['China']
listofyear = ['2020']
imports = df.loc[df['Indicator'].isin(list1)]
imports = imports.loc[imports['Reporting Economy'].isin(listofcountry)]
imports = imports.loc[imports['Year'].isin(listofyear)]
```

```python
imports
```

```python
imports_China = imports[['Indicator','Period','Value']]
```

```python
imports_China
```

```python
plt.rcParams["figure.figsize"] = (10, 10)

plt.plot(imports_China['Period'],imports_China['Value'], color='darkslategrey')

plt.xlabel('Months for the Year 2020')
plt.ylabel('Total Merchandise Imports in Millions of US Dollars')
plt.ylim(140000, 300000)
plt.fill_between(imports_China['Period'],imports_China['Value'], color = 'darkcyan')
plt.title('Total Merchandise Imports of China with the World in the Year 2020')
plt.grid(True)
plt.savefig('importChina.png')
plt.show()
```
