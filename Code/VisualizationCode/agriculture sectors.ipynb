```python
plt.rcParams["figure.figsize"] = (5, 4)

plt.plot(df['Month'],df['Agriculture'])

plt.xlabel('Months')
plt.ylabel('Active Business Percentage')
plt.ylim(90, 100)
plt.title('Agriculture Sector')
plt.grid(True)
plt.fill_between(df['Month'],df['Agriculture'], color = 'lightsteelblue')

plt.savefig('agriculture.png')

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (5, 4)

plt.plot(df['Month'],df['PS'], color = 'salmon')

plt.xlabel('Months')
plt.ylabel('Active Business Percentage')
plt.ylim(10,100)
plt.title('Personal Services Sector')
plt.grid(True)
plt.fill_between(df['Month'],df['PS'], color = 'peachpuff')

plt.savefig('ps.png')

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (5, 4)


plt.plot(df['Month'],df['Construction'], color = 'orange')

plt.ylim(73, 93)

plt.xlabel('Months')
plt.ylabel('Active Business Percentage')
plt.fill_between(df['Month'],df['Construction'], color = 'peach')
plt.title('Construction Sector')
plt.grid(True)


plt.show()
```
